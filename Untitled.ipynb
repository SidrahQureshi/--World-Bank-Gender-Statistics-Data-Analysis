# World Bank Gender Statistics Data Analysis

In this notebook, I analyse the World Bank Gender Statistics.
The python libraries I use in this analysis include pandas, numpy, matplotlib and seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


gender_df = pd.read_csv('Gender_Statistics_Data.csv')
poverty_df=pd.read_csv('Poverty_and_Equity_Data.csv')
gender_df.head()
print(gender_df.shape[0])
print(poverty_df.shape[0])
print(poverty_df.describe())
```

### Data Cleaning

The Gender Statistics Data is from year 1971 onwards while the Poverty and Equity data is for the last 20 years.
We are going to analyse the data for the last 20 years data, therefore, we can delete all the other year columns from the Gender data.

```python
gender_df = gender_df.drop(
    ['1971 [YR1971]', '1972 [YR1972]','1973 [YR1973]','1974 [YR1974]','1975 [YR1975]','1976 [YR1976]','1977 [YR1977]','1978 [YR1978]','1979 [YR1979]',
    '1980 [YR1980]','1981 [YR1981]','1982 [YR1982]','1983 [YR1983]','1984 [YR1984]','1985 [YR1985]','1986 [YR1986]','1987 [YR1987]','1988 [YR1988]','1989 [YR1989]',
     '1990 [YR1990]','1991 [YR1991]','1992 [YR1992]','1993 [YR1993]','1994 [YR1994]','1995 [YR1995]','1996 [YR1996]','1997 [YR1997]','1998 [YR1998]','1999 [YR1999]','2020 [YR2020]'], # Labels of columns to drop
    axis = 1 # Drop columns (axis=1), not rows (axis=0)
)
```

2- Looking at the data, we can see that the year columns contain the characters '..' where the data is not present. We are going to replace '..' with NaN so we use those columns in our calculations

```python

gender_stats_cols=['2000 [YR2000]','2001 [YR2001]','2002 [YR2002]','2003 [YR2003]','2004 [YR2004]','2005 [YR2005]','2006 [YR2006]','2007 [YR2007]'
     ,'2008 [YR2008]','2009 [YR2009]','2010 [YR2010]','2011 [YR2011]','2012 [YR2012]','2013 [YR2013]','2014 [YR2014]','2015 [YR2015]'
     ,'2016 [YR2016]','2017 [YR2017]','2018 [YR2018]','2019 [YR2019]']
poverty_stats_cols=['2000 [YR2000]','2001 [YR2001]','2002 [YR2002]','2003 [YR2003]','2004 [YR2004]','2005 [YR2005]','2006 [YR2006]','2007 [YR2007]'
     ,'2008 [YR2008]','2009 [YR2009]','2010 [YR2010]','2011 [YR2011]','2012 [YR2012]','2013 [YR2013]','2014 [YR2014]','2015 [YR2015]'
     ,'2016 [YR2016]','2017 [YR2017]','2018 [YR2018]','2019 [YR2019]']
year_2010_2019=['2010 [YR2010]','2011 [YR2011]','2012 [YR2012]','2013 [YR2013]','2014 [YR2014]','2015 [YR2015]'
     ,'2016 [YR2016]','2017 [YR2017]','2018 [YR2018]','2019 [YR2019]']
year_2000_2009=['2000 [YR2000]','2001 [YR2001]','2002 [YR2002]','2003 [YR2003]','2004 [YR2004]','2005 [YR2005]','2006 [YR2006]','2007 [YR2007]'
     ,'2008 [YR2008]','2009 [YR2009]']

for col in gender_stats_cols:
    gender_df[gender_stats_cols] = gender_df[gender_stats_cols].replace('..',np.NaN)
for col in poverty_stats_cols:
    poverty_df[poverty_stats_cols] = poverty_df[poverty_stats_cols].replace('..',np.NaN)

print(gender_df)

```

Since the rows with no values in the year columns are of no use to us, we can drop them from the dataframe.

```python
#gender_df[cols] = gender_df[cols].fillna(value=gender_df.mean(axis=1))
gender_df=gender_df.dropna(axis=0, how='all', subset=gender_stats_cols)
poverty_df=poverty_df.dropna(axis=0, how='all', subset=poverty_stats_cols)

#gender_df[cols] = gender_df[cols].fillna(value=gender_df[cols].mean(axis=1))


print(gender_df.shape[0])

print(gender_df.dtypes)
```

Convert datatype of the year columns from object to float

```python
gender_df[gender_stats_cols]=gender_df[gender_stats_cols].astype(float)
poverty_df[poverty_stats_cols]=poverty_df[poverty_stats_cols].astype(float)
```

Fill the NaN values with mean of the values against all the years in the data

```python
gender_df[gender_stats_cols] = gender_df[gender_stats_cols].fillna(gender_df[gender_stats_cols].mean(axis=1)[0])
print(poverty_df.head(10))
#poverty_df[poverty_stats_cols] = poverty_df[poverty_stats_cols].fillna(poverty_df[poverty_stats_cols].mean(axis=1)[0])

#print(gender_df.head(10))
#print(poverty_df.head(10))
```

We are now in the position of analysing our data for some answers and patterns.

#### The main decision maker about a woman's visits to her family or relatives

Series code for series 'Decision maker about a woman's visits to her family or relatives: mainly wife (% of women age 15-49)'
is SG.DMK.VISI.WF.ZS

```python
decision_maker_df=gender_df[(gender_df['Series Code']=='SG.DMK.VISI.SE.ZS')
                            |(gender_df['Series Code']=='SG.DMK.VISI.WF.ZS')
                            |(gender_df['Series Code']=='SG.DMK.VISI.HB.ZS')
                            |(gender_df['Series Code']=='SG.DMK.VISI.WH.ZS')
                            |(gender_df['Series Code']=='SG.DMK.VISI.OT.ZS')]

decision_maker_wf=gender_df[(gender_df['Series Code']=='SG.DMK.VISI.WF.ZS')]
decision_making_mean=decision_maker_wf[gender_stats_cols].mean(axis=1).sort_values(
    ascending = True
)
fig = plt.figure()

fig.set_size_inches(15,10)
plt.xlabel('Countries')
plt.ylabel("Percentage of Women who have the decision making power about visits to family")
plt.xticks(rotation='vertical')
bar_graph = plt.bar(decision_maker_wf['Country Name'],decision_making_mean )
plt.savefig('plot1.png')
plt.figure()
```

vertical bar chart showing female employment in agriculture(SL.AGR.EMPL.FE.ZS), industry, service (SL.SRV.EMPL.FE.ZS) ..by region
Heavily indebted poor countries (HIPC)	HPC
High income	HIC

```python
female_participation_df=gender_df[(gender_df['Series Code']=='SL.SRV.EMPL.FE.ZS')
                            |(gender_df['Series Code']=='SL.AGR.EMPL.FE.ZS')
                            |(gender_df['Series Code']=='SL.IND.EMPL.FE.ZS')]

region_subset_df=female_participation_df[(female_participation_df['Country Code']=='HPC')
                                          |(female_participation_df['Country Code']=='HIC')
                                        |(female_participation_df['Country Code']=='FCS')]

region_selected_col_df = region_subset_df[['Series Code','Country Code','Country Name', '2019 [YR2019]']]


fig, ax = plt.subplots(figsize=(12, 8))
x = np.arange(len(region_selected_col_df['Country Name'].unique()))
bar_width = 0.2
b1 = ax.bar(x, region_selected_col_df.loc[region_selected_col_df['Series Code'] == 'SL.SRV.EMPL.FE.ZS', '2019 [YR2019]'],
            width=bar_width, label='Services')
b2 = ax.bar(x + bar_width, region_selected_col_df.loc[region_selected_col_df['Series Code'] == 'SL.AGR.EMPL.FE.ZS', '2019 [YR2019]'],
            width=bar_width, label='Agriculture')
b3 = ax.bar(x + bar_width*2, region_selected_col_df.loc[region_selected_col_df['Series Code'] == 'SL.IND.EMPL.FE.ZS', '2019 [YR2019]'],
            width=bar_width, label='Industry')
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(region_selected_col_df['Country Name'].unique())
ax.legend()
ax.set_ylabel('Employment (%)', labelpad=15)
ax.set_title('Female Employement Percentage in Agriculture, Services and Industry by Region', pad=15)

```

#### Comparison between the countries with and without gender descrimination laws

Which countries have laws that prohibit gender discrimination in workplaces?
The indicator for gender discrimination laws is "Law prohibits discrimination in employment based on gender (1=yes; 0=no)" with Series code value : SG.LAW.NODC.HR

```python
#Countries with laws against gender discriminiation
gender_law_yes_df=gender_df[(gender_df['Series Code']=='SG.LAW.NODC.HR')&(gender_df['2019 [YR2019]']==1.0)]
print("Countries with laws against gender discriminiation")
print(gender_law_yes_df['Country Name'])

#Countries with no laws against gender discriminiation
gender_law_no_df=gender_df[(gender_df['Series Code']=='SG.LAW.NODC.HR')&(gender_df['2019 [YR2019]']==0.0)]
print("Countries without laws against gender discriminiation")
print(gender_law_no_df['Country Name'])
```

How is female employment percentage related to having laws that prohibit gender discrimination?

```python
labor_law_df=gender_df[(gender_df['Series Code']=='SL.TLF.TOTL.FE.ZS')]
#declaring empty dataframes
labor_no_law=gender_df[(gender_df['Series Code']=='SL.TLF.TOTT.FE.ZS')]
labor_yes_law=gender_df[(gender_df['Series Code']=='SL.TLF.TOTT.FE.ZS')]

#labor_yes_law=labor_yes_law[(labor_yes_law['2019 [YR2019]'].quantile([0.0]))]


for country in labor_law_df['Country Code']:
    if (gender_law_no_df['Country Code'].isin([country])).any():
      labor_no_law= labor_no_law.append(labor_law_df[labor_law_df['Country Code']==country], ignore_index=True)
    else:
      labor_yes_law= labor_yes_law.append(labor_law_df[labor_law_df['Country Code']==country], ignore_index=True)


fig = plt.figure(figsize =(20, 30))

fig.set_size_inches(20,20)
plt.ylabel('Countries with gender discrimination laws')
plt.xlabel("Percentage of Women in labour force")
plt.yticks(rotation='horizontal')
bar_graph = plt.barh(labor_yes_law['Country Name'],labor_yes_law['2019 [YR2019]'].sort_values(
    ascending = True) )
plt.figure(figsize =(20, 30)) 
plt.savefig('plot3.png')
fig = plt.figure()

fig.set_size_inches(20,30)
plt.ylabel('Countries without gender discrimination laws')
plt.xlabel("Percentage of Women in labour force")
plt.yticks(rotation='horizontal')
bar_graph2 = plt.barh(labor_no_law['Country Name'],labor_no_law['2019 [YR2019]'].sort_values(
    ascending = True
) )
plt.savefig('plot4.png')
plt.figure() 

 
```
